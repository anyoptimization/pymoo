## BNH

Binh and Korn defined the following test problem in <cite data-cite="bnh"></cite> with 2 objectives and 2 constraints:

**Definition**

\begin{align}
\begin{split}
f_1(x) & =\, & 4x_1^2 + 4x_2^2\\
f_2(x) & = & (x_1-5)^2 + (x_2-5)^2\\[2mm]
g_1(x) & = & (x_1-5)^2 + x_2^2 \leq 25 \\
g_2(x) & = & (x_1-8)^2 + (x_2+3)^2 \geq 7.7\\[2mm]
\end{split}
\end{align}

$$ 0 \leq x_1 \leq 5 \\
0 \leq x_2 \leq 3$$

**Optimum**

$$x^*_1 = x^*_2 \; \in \; [0,3] \quad \text{and} \quad  x^*_1  \in [3,5],\; x^*_2 = 3$$

**Plot**

```python
from pymoo.factory import get_problem
from pymoo.util.plotting import plot

problem = get_problem("bnh")
plot(problem.pareto_front(), no_fill=True)
```
